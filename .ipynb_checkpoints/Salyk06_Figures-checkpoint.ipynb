## Salyk et al. 2006, Figure 5, top

```python
fig=plt.figure(figsize=(7,4))
ax1=fig.add_subplot(111)
ax1.plot(pdata['lat'],pdata['upvpbar']*udata['dudy']*1e4,'k',label="u'v' x du/dy x 10$^4$")
ax1.set_xlabel('Planetocentric Latitude [degrees]')
ax1.legend()
ax1.set_xlim(-60,60)
ax1.set_ylim(-2,10)
ax1.grid()
```

## Salyk et al. 2006, Figure 5, bottom

```python
fig=plt.figure(figsize=(7,4))
ax1=fig.add_subplot(111)
ax1.plot(pdata['lat'],pdata['upvpbar'],'k.',label="u'v' [m$^2$s$^{-2}$]")
ax1.errorbar(pdata['lat'],pdata['upvpbar'],yerr=pdata['sigupvpbar']*2.,fmt='none',color='k')
ax1.plot(udata['lat'],udata['dudy']*1e6,label='du/dy x 10$^6$ [s$^{-1}$]')
ax1.set_xlabel('Planetocentric Latitude [degrees]')
ax1.legend()
ax1.set_xlim(-60,60)
ax1.set_ylim(-50,50)
ax1.grid()
```
